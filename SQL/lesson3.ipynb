```python
#основные элементы релиационных Баз Данных 
# Таблицы связанные друг с другом с помощью ключевых полей, эти поля называются первичным ключом и внешним ключом.
# Каждая таблица в бд содержит первичный ключ. 
# Первичный ключ - уникальный. И он НЕ может быть Нулевым (NULL) 
В БД таблицы будем называть связи 
Связи в БД имеет несколько видов. 

```

Связи делятся на:

1 Многие ко многим.
2 Один ко многим.
-с обязательной связью;
-с необязательной связью;
3 Один к одному.
-с обязательной связью;
-с необязательной связью;

Первичный ключ - он всегда находится в первом столбце
Внешний ключ - это поле в БД, он обычно находится гдето во 2 или 3 столбце. Где внешний ключ соотвествует первичному ключу в другой таблице. Таким образом создается связь.

схема "сущность-связь"

- Когда Символ 1 и бесконечность, это значит связь "один ко многим" 
- Когда бесконечность и бесконечность, это значит связь "многие ко многим"
