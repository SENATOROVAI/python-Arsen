```python
#arbuz 
# https://codeforces.com/problemset/problem/4/A input - 8
print('YNEOS'[5%~int(input())%2::2])
```

int(input()): Эта строка считывает ввод пользователя и преобразует его в целое число.

~int(input())%2: Эта строка инвертирует биты числа, полученного из ввода, и вычисляет остаток от деления на 2. Если число четное, то результат будет 0, иначе -1.

5%~int(input())%2: Эта строка вычисляет остаток от деления числа 5 на результат из предыдущего шага. Если остаток от деления на 2 был 0 (число четное), то результат будет 5. Если остаток от деления на 2 был -1 (число нечетное), то результат будет 4.

'YNEOS'[5%~int(input())%2::2]: Эта строка использует полученное число в качестве индекса для строки 'YNEOS'. Если индекс будет 4, то результат будет 'E', если индекс будет 5, то результат будет 'O'.

print(...): Эта строка выводит результат на экран.

В итоге, если введенное число четное, выведется 'Y', если нечетное - 'N'.

Однако, это решение не является читаемым и не рекомендуется использовать в реальных проектах, так как оно использует нестандартные конструкции и может быть трудно понять, что оно делает. Лучше использовать более обычный и понятный код, как в примерах выше.

```python
w = int(input()) - 2
print("YES" if w > 0 and w % 2 == 0 else "NO")
```

w = int(input()) - 2


1. `int(input())`: Эта строка считывает ввод пользователя и преобразует его в целое число.

2. `- 2`: Эта строка вычитает 2 из полученного числа. Это сделано для того, чтобы учесть вес арбуза без учета двух килограммов, которые уже были взяты Петей и Васей.


print("YES" if w > 0 and w % 2 == 0 else "NO")


3. `w > 0`: Эта проверка проверяет, что остаток веса после вычитания 2 больше 0, чтобы гарантировать, что остаток веса после взятия двух килограммов положителен.

4. `w % 2 == 0`: Эта проверка проверяет, что остаток веса после вычитания 2 четный.

5. `"YES" if ... else "NO"`: Это условный оператор, который выводит "YES", если оба условия выполняются (остаток веса больше 0 и четный), иначе выводит "NO".

В итоге, если остаток веса после взятия двух килограммов четный и больше 0, выведется 'YES', иначе - 'NO'. Это решение более читаемо и понятно, чем первое решение, предложенное выше.
