```python
# Куплю лопату https://codeforces.com/contest/732/problem/A 

k, r=map(int,input().split())
l=0
su=0
p=0
while l==0:
    p+=1
    su+=k
    if su%10==r or su%10==0:
        l=1
        print(p)
        exit()
```

```python
k, r = [int(x) for x in input().split(' ')]
for i in range(1, 10):
    if (k * i) % 10 == r:
        print(i)
        break
    else:
        if (k * i) % 10 == 0:
            print(i)
            break
```

```python
p, r=map(int, input().split())
print(min(i for i in range(1, 11) if (i*p)%10 in [0,r]))
```
